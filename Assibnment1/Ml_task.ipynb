```python
from sklearn.linear_model import LinearRegression

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error

```

```python
import pandas as pd 
df = pd.read_csv("tips.csv")
df.head()
```

```python
df.isnull()
```

```python
df.columns
```

```python

from sklearn.preprocessing import LabelEncoder



# Columns to encode
en_columns = ['sex','smoker','day','time']

label_encode = {}

for col in en_columns:
    le = LabelEncoder()                
    df[col] = le.fit_transform(df[col])  
    label_encode[col] = le 

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = df[['total_bill']]
y = df['tip']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


lr = LinearRegression()
lr.fit(x_train, y_train)


pre_y = lr.predict(x_test)


print("R2 Score for Linear Regression: ", r2_score(y_test, pre_y))
print("MAE for Linear Regression: ", mean_absolute_error(y_test, pre_y))
print("MAPE for Linear Regression: ", mean_absolute_percentage_error(y_test, pre_y))
```

```python
import pickle
with open("model.pkl", "wb") as file:
    pickle.dump(lr, file)
```
